```python
import pandas as pd
from collections import OrderedDict

def get_timing_info(filename, casename=None):
    if casename is None:
        casename = filename
    
    keys = ['GEOS-Chem', 'HEMCO', 'All chemistry', 'Transport', 'Convection', 'Boundary layer mixing']
    timing_info = OrderedDict()
    timing_info['case'] = casename

    with open(filename, "r") as f:
        for line in f:
            for key in keys:
                if line.startswith("  "+key):
                    timing_info[key] = line.split()[-1]
    
    return timing_info
```

```python
import os
import glob

#filenames = ['./AWS/'+name for name in os.listdir("./AWS")]
#filenames = glob.glob("./*/timing_7days_*.log")
filenames = glob.glob("./*/timing_7days_*rerun.log")
filenames
```

```python
#casenames = [filename.split('/')[-1][13:-4] for filename in filenames]
casenames = [filename.split('/')[-1][13:-10] for filename in filenames]
casenames
```

```python
time_info_list = [get_timing_info(filename, casename) for filename, casename in zip(filenames, casenames)]
```

```python
pd.DataFrame.from_records(time_info_list)
```
